```python

import nltk
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('customer_reviews.csv')
print("First 5 rows of the dataset:")
print(df.head())
```

```python
# 2. Text Preprocessing (20 Points)
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Initialize stop words and lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
# Define text preprocessing function
def preprocess_text(text):
    # Remove special characters
    text = re.sub(r'[^A-Za-z0-9 ]+', '', text)
    # Convert to lowercase
    text = text.lower()
    # Tokenize text
    tokens = word_tokenize(text)
    # Remove stop words and lemmatize each token
    tokens = [lemmatizer.lemmatize(t) for t in tokens if t not in stop_words]
    # Return cleaned text
    return ' '.join(tokens)
```

```python
# Apply preprocessing to the review_text column
df['cleaned_text'] = df['review_text'].apply(preprocess_text)
print("\nFirst 5 rows with cleaned text:")
print(df[['review_text', 'cleaned_text']].head())
```

```python
# 3. Exploratory Data Analysis (EDA) on Text Length (10 Points)
# Calculate length of each cleaned review
df['review_length'] = df['cleaned_text'].apply(len)
```

```python
# Plot histogram of review lengths
plt.figure(figsize=(10, 6))
sns.histplot(df['review_length'], kde=True, color='skyblue')
plt.title('Distribution of Review Text Lengths')
plt.xlabel('Review Length')
plt.ylabel('Frequency')
plt.show()
```
